```python

# comments
import time
import pandas as pd
from rich import print
from rich.console import Console
from rich.table import Table
from rich import box
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
time.sleep(0.3)
console = Console()

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "score": [88.5, 92.1, 79.3],
    "city": ["NYC", "LA", "Chicago"],
    "active": [True, False, True],
})

table = Table(
    title="📊 User Performance Dashboard",
    box=box.ROUNDED,
    header_style="bold magenta",
    show_lines=True,
)

table.add_column("Name", style="cyan", justify="left")
table.add_column("Age", justify="right")
table.add_column("Score", justify="right")
table.add_column("City", style="green")
table.add_column("Active", justify="center")

for _, row in df.iterrows():
    score_style = "bold green" if row["score"] >= 90 else "yellow"
    active_style = "green" if row["active"] else "red"

    table.add_row(
        row["name"],
        str(row["age"]),
        f"[{score_style}]{row['score']:.1f}[/{score_style}]",
        row["city"],
        f"[{active_style}]{row['active']}[/{active_style}]",
    )

console.print(table)
```
